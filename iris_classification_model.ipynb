# 제목: 꽃의 특징으로 품종을 맞히는 모델 만들기


목표
1. Python을 이용한 AI 모델 만들기
2. 데이터를 이용해 학습(Machine Learning) 하고 예측(Prediction)하는 과정을 이해하기
3. 실제 데이터를 입력해서 확인

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from IPython.display import Image # 이미지용
```

데이터 소개 (붓꽃 - iris)
품종 3가지: Setosa(세토사), Versicolor(버시컬러), Virginica(버지니카)

각 꽃은 다음 4가지 특성
1. 꽃받침 길이(sepal length)
2. 꽃받침 너비(sepal width)
3. 꽃잎 길이(petal length)
4. 꽃잎 너비(petal width)

```python
Image('https://user-images.githubusercontent.com/15958325/56006707-f69f3680-5d10-11e9-8609-25ba5034607e.png')

```

```python
#step 1. 데이터 불러오기
iris = load_iris()
X = iris.data
Y = iris.target
#print(X)
print("___________________________________________________________________")
#print(Y)
#print(len(X),len(Y))
#데이터 display
pd.DataFrame(X, columns=iris.feature_names).head()
```

```python
#Step 2: AI가 공부할 데이터 준비
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#Step 3: AI 모델 만들기
model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

```python
#Step 4: AI가 테스트를 보게 하기
Y_pred = model.predict(X_test)
acc = accuracy_score(Y_test, Y_pred)
print('Accuracy:', acc*100,'%')
```

```python
#from sys import setdlopenflags
#Step 5: 직접 꽃 크기 입력(sepal, petal)
#sample data
#5.1 3.5 1.4 0.2 --> setosa
#6.0 2.2 4.0 1.0 --> versicolor
#7.7 3.0 6.1 2.3 --> virginica
print('꽃의 크기를 입력해 주세요!')
sl = float(input('꽃받침 길이(sepal length): '))
sw = float(input('꽃받침 길이(sepal width): '))
pl = float(input('꽃잎 길이(sepal length): '))
pw = float(input('꽃잎 길이(sepal width): '))

user_input = [[sl,sw,pl,pw]]
prediction = model.predict(user_input)
print('Prediction data value: ', prediction)
print("네 예측된 품종은: ", iris.target_names[prediction[0]])
```
